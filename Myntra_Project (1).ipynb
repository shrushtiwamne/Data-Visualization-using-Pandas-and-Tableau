```python
# lets work on DiscountPrice (in Rs)  and DiscountOffer columns for nulls
# Its is appropriate to  replace nulls in Discount offer with 0 if there is no discount to offer.
df["DiscountOffer"] = df["DiscountOffer"].fillna(0)
df.head(20)
```

```python
# Now we should deal with DiscountPrice (in Rs) columns since there are nulls even though discount offer is not null

#lest split discount offer where % and Rs are there for better "discount in price"

df["discount_in_per"]=df["DiscountOffer"].apply(lambda x: x.split('%')[0] if isinstance(x, str) and '%' in x else 0)
df.head(5)
```

```python
# split with Rs in it
df["discount_in_rs"]= df["DiscountOffer"].apply(lambda x : x.split(" ")[1] if isinstance(x, str) and "Rs. " in x else 0)
df.tail(50)
```

```python
df["discount_in_rs"] = df["discount_in_rs"].astype(float)
df["discount_in_per"] = df["discount_in_per"].astype(float)
```

```python
# Based on above two newly created colums we can find discounted price

#lets calculate discount in price using newly created columns
df["New_Discount_price"] = df.apply(lambda x: (x["OriginalPrice (in Rs)"] * x["discount_in_per"]) / 100 if x["discount_in_per"] != 0 and x["discount_in_rs"] == 0 else  x["discount_in_rs"] if x["discount_in_per"] == 0 and x["discount_in_rs"] != 0 else 0, axis=1)
df.tail(50)
```

```python
# we must calculated new discounted price from original price)
df["New_Discounted_price"]=df["OriginalPrice (in Rs)"]-df["New_Discount_price"]

df.tail(50)
```

```python
df.drop(columns=["discount_in_per", "discount_in_rs","URL","Description","DiscountPrice (in Rs)"], inplace=True)
```

```python
#Renaming few columns for better understanding
df.rename(columns={
    "category_by_Gender": "Gender",
    "New_Discounted_price": "Discounted_Price",
    "OriginalPrice (in Rs)": "OriginalPrice",
    "New_Discount_price": "Discount_In_Rs"
}, inplace=True)
df.head()
```
